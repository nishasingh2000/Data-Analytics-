# Statistic Basics

### 1. What is statistics, and why is it important ?
    - Statistics is the branch of mathematics that deals with the collection, analysis, interpretation, presentation, and organization of data. It helps us understand and describe population in the world around us through numbers and data.
    It is important because it helps us for decision-making, understanding Trends and patterns, Quality Control & Risk Assessment.
### 2. What are the two main types of statistics ?
    - Statistics main two are types are: Descriptive Statistics and Inferential Statistics
### 3. What are descriptive statistics ?
    - Descriptive statistics is a type of statistics in which we'll describing the data and make decisions according to the data. for e.g Taking the average height of students in Class X.
### 4. What is inferential statistics ? 
    - Inferential statistics is a 2nd type of statistics in which we'll take an sample from the population and predicting the values for population according to the results that we got in sample data.
### 5. What is sampling in statistics ?
    - Sampling in statistics is the process of selecting a subset (called a sample) from a larger group (called a population) to analyze and draw conclusions about the whole population.
### 6. What are the different types of sampling methods?
    - The different types of sampling methods are as given below:-
    1. Random Sampling: Every member has an equal chance of being selected.
    2. Systematic Sampling: Selecting every kth individual (e.g., every 10th person).
    3. Stratified Sampling: Dividing the population into subgroups and sampling from each. 
    4. Cluster Sampling: Dividing the population into clusters and randomly selecting entire clusters. 
### 7. What is the difference between random and non-random sampling?
    - Random Sampling: Every member of the population has an equal chance of being selected. For example: Drawing names from a hat.
    - Non-Random Sampling: Selection is based on convenience or judgment, not chance.For example: Interviewing only people at a specific location.
### 8. Define and give examples of qualitative and quantitative data.
    - Qualitative Data: Descriptive, non-numerical information. For example: Colors of cars, types of cuisine.
    - Quantitative Data: Numerical information. For example: Age, weight, temperature.
### 9. What are the different types of data in statistics ?
    - Categorical (Qualitative): Nominal, Ordinal.
    - Numerical (Quantitative): Discrete, Continuous.
### 10. Explain nominal, ordinal, interval, and ratio levels of measurement.
    - Nominal: Categories without order. For example: Gender, eye color.
    - Ordinal: Categories with order. For example: Survey ratings (good, better, best).
    - Interval: Numeric, equal intervals, no true zero. For example: Temperature in Celsius.
    - Ratio: Numeric, equal intervals, with true zero. For example: Height, weight, age.
### 11. What is the measure of central tendency ?
    - It is a type of descriptive statistics which is used to find the center of the data. It includes:- Mean, Median, Mode.
### 12. Define mean, median, and mode.
    - Mean:- Average Value in the data
    - Median:- Middle Most value in the sorted data.
    - Mode:- Most frequent value in the data.
### 13. What is the significance of the measure of central tendency ?
    - It helps summarize a large dataset with a single representative value for easy comparison and understanding.
### 14. What is variance, and how is it calculated ?
    - Variance measures how far data points spread out from the mean. It is calculated by:- Variance = Summation(x-x_bar)^2/n Where x = each value, x_bar = mean, n = number of values
### 15. What is standard deviation, and why is it important ?
    - Standard deviation is the square root of variance; shows data spread in original units. It helps understand data consistency and variability.
### 16. Define and explain the term range in statistics.
    - Range is defined as the difference between the highest and lowest value. Range = Max - Min
### 17. What is the difference between variance and standard deviation ?
    - Variance is the Average squared deviation. and Standard deviation the the square root of variance
### 18. What is skewness in a dataset ?
    - Skewness measures asymmetry in a distribution. It;'s of two types:- Positive Skew: Tail on the right & Negative Skew: Tail on the left.
### 19. What does it mean if a dataset is positively or negatively skewed ?
    - Postive skew means :- Mean > Median > Mode
    - Negative skew means:- Mean < Median < Mode
### 20. Define and explain kurtosis.
    - kurtosis measures the "tailedness" or peakedness of a distribution. It is of three types:- 
    1. Leptokurtic: Sharp peak.
    2. Platykurtic: Flat peak.
    3. Mesokurtic: Normal peak.
### 21. What is the purpose of covariance ?
    - Covariance shows how two variables change together.
    1. Positive covariance: Both increase/decrease together.
    2. Negative Covariance: One Increases, the other decreases.
### 22. What does correlation measure in statistics ?
    - Correlation measures strength and direction of a linear relationship between two variables. It's Range is between -1 to +1
### 23. What is the difference between covariance and correlation ?
    - Covariance is unscaled and it is unit-dependent.
    - Correlation is scaled(Unitless), easier to interpret.
### 24. What are some real-world applications of statistics ?
    - The real-world applications of statistics is as given below:-
    1. Healthcare:- Drug effectiveness analysis.
    2. Business:- Market research and forecasting.
    3. Education:- Exam score analysis

```python
# 4. Implement sampling techniques like random sampling and stratified sampling.
import random
def random_sampling(data, sample_size):
    return random.sample(data, sample_size)
def stratified_sampling(data, strata, sample_size):
    stratified_samples = []
    for stratum in strata:
        stratum_data = [item for item in data if item['stratum'] == stratum]
        stratified_samples.extend(random.sample(stratum_data, sample_size))
    return stratified_samples
data = [
    {'value': 1, 'stratum': 'A'},
    {'value': 2, 'stratum': 'A'},
    {'value': 3, 'stratum': 'B'},
    {'value': 4, 'stratum': 'B'},
    {'value': 5, 'stratum': 'C'},
    {'value': 6, 'stratum': 'C'},
]
sample_size = 1
random_sample = random_sampling(data, sample_size)
strata = ['A', 'B', 'C']
stratified_sample = stratified_sampling(data, strata, sample_size)
print(f"Random Sample: {random_sample}")
print(f"Stratified Sample: {stratified_sample}")
```

```python
# 9. Write a Python script to calculate covariance between two datasets.
def calculate_covariance(data1, data2):
    if len(data1) != len(data2):
        raise ValueError("Datasets must be of the same length")
    mean1 = statistics.mean(data1)
    mean2 = statistics.mean(data2)
    covariance = sum((x - mean1) * (y - mean2) for x, y in zip(data1, data2)) / len(data1)
    return covariance
data1 = [1, 2, 3, 4, 5]
data2 = [5, 4, 3, 2, 1]
covariance = calculate_covariance(data1, data2)
print(f"Covariance: {covariance}")
```

```python
# 10. Write a Python script to calculate the correlation coefficient between two datasets.
def calculate_correlation(data1, data2):
    if len(data1) != len(data2):
        raise ValueError("Datasets must be of the same length")
    mean1 = statistics.mean(data1)
    mean2 = statistics.mean(data2)
    covariance = calculate_covariance(data1, data2)
    std_dev1 = statistics.stdev(data1)
    std_dev2 = statistics.stdev(data2)
    correlation = covariance / (std_dev1 * std_dev2)
    return correlation
data1 = [1, 2, 3, 4, 5]
data2 = [5, 4, 3, 2, 1]
correlation = calculate_correlation(data1, data2)
print(f"Correlation Coefficient: {correlation}")
```

```python
# 13. Calculate the mean, median, and mode of grouped data.
def grouped_data_statistics(data):
    total_frequency = sum(data.values())
    mean = sum(value * frequency for value, frequency in data.items()) / total_frequency
    median = None
    cumulative_frequency = 0
    for value, frequency in sorted(data.items()):
        cumulative_frequency += frequency
        if cumulative_frequency >= total_frequency / 2:
            median = value
            break
    mode = max(data, key=data.get)
    return mean, median, mode
grouped_data = {1: 2, 2: 3, 3: 5, 4: 4}
mean, median, mode = grouped_data_statistics(grouped_data)
print(f"Grouped Data - Mean: {mean}, Median: {median}, Mode: {mode}")
```

```python
# 17. Calculate the interquartile range (IQR) of a dataset.
def calculate_iqr(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    return iqr
data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
iqr = calculate_iqr(data)
print(f"Interquartile Range (IQR): {iqr}")
```

```python
# 18. Implement Z-score normalization and explain its significance.
def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return [(x - mean) / std_dev for x in data]
data = [1, 2, 3, 4, 5]
normalized_data = z_score_normalization(data)
print(f"Z-score Normalized Data: {normalized_data}")
```

```python
# 19. Compare two datasets using their standard deviations.
def compare_standard_deviations(data1, data2):
    std_dev1 = np.std(data1)
    std_dev2 = np.std(data2)
    return std_dev1, std_dev2
data1 = [1, 2, 3, 4, 5]
data2 = [5, 6, 7, 8, 9]
std_dev1, std_dev2 = compare_standard_deviations(data1, data2)
print(f"Standard Deviation of Data1: {std_dev1}")
print(f"Standard Deviation of Data2: {std_dev2}")
```
