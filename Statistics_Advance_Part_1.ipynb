### 1. What is a random variable in probability theory ?
    - A random variable is a function that assigns a numerical value to each outcome in a sample space of a random experiment.
### 2. What are the types of random variables ?
    - There are two types of random variables:
    1. Discrete Random Variable: It takes on a countable number of values. for e.g:- Number on a rolled die (1-6)
    2. Continuos Random Variable: Takes on infinitely many values within an interval. for e.g:- Time taken to complete a task (e.g., 1.5 seconds, 2.7 seconds)
### 3. What is the difference between discrete and continuous distributions ?
    - The difference between discrete and continuous distributions is that as given below:-
    1. A discrete distribution is associated with a discrete random variable, which can take on a countable number of distinct values.
    2. A continuous distribution is associated with a continuous random variable, which can take on infinitely many values in an interval (e.g., any real number between 0 and 1).
### 4. What are probability distribution functions (PDF) ?
    - In probability theory, a Probability Distribution Function (PDF) describes how the values of a continuous random variable are distributed‚Äîthat is, how likely the variable is to take on a value within a given range.
### 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF) ?
    - PDF (Probability Density Function):- It is used only for continous random variables. It tells you how dense the probability is at a specific value.
    - CDF (Cumulative Distribution Function):- It is used for both discrete and continuous variables. It gives the probability that the variable is less than or equal to a value.
### 6. What is a discrete uniform distribution ?
    - A Discrete Uniform Distribution is a type of probability distribution in which all outcomes are equally likely over a finite set of distinct values. for e.g rolling a fair 6-sided die: Possible values are (1,2,3,4,5,6) & Probability of each outcome is (1/6)
### 7. What are the key properties of a Bernoulli distribution ?
    - The Bernoulli distribution is one of the simplest and most fundamental discrete probability distributions. It models a random experiment that has only two possible outcomes: Success (1) and Failure (0)
### 8. What is the binomial distribution, and how is it used in probability ?
    - The Binomial Distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials (i.e., yes/no experiments), each with the same probability of success. For e.g testing 20 light bulbs and counting how many are defective.
### 9. What is the Poisson distribution and where is it applied ?
    - The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space, given that: The events occur independently, at a constant average rate, and not simultaneously.
### 10. What is a continuous uniform distribution ?
    - A Continuous Uniform Distribution is a type of probability distribution where all values in a continuous interval are equally likely. It is also known as a rectangular distribution because its probability density function (PDF) forms a rectangle.
### 11. What are the characteristics of a normal distribution ?
    - The Normal distribution (also called Gaussian distribution) is one of the most important continuous probability distributions in statistics. It is symmetric and describes many natural phenomena.
### 12. What is the standard normal distribution, and why is it important ?

### 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics ?
    - The Central Limit Theorem (CLT) is a fundamental concept in statistics that states:- When you take large enough samples from any population (regardless of the population's distribution), the sampling distribution of the sample mean will approximate a normal distribution.
    - It is critical because:- CLT justifies why many statistical procedures assume normality for sample means, allowing us to use normal distribution tools (z-scores, t-tests, confidence intervals) even if the population isn't normal.
    - It allows approximation of probabilities related to sample means with normal distribution tables, even when the population distribution is unknown or complicated.
### 14. How does the Central Limit Theorem relate to the normal distribution ?
    - CLT explains how and why the normal distribution arises naturally in the context of averages of many random variables.
### 15. What is the application of Z statistics in hypothesis testing ?
    - Z-statistics (or Z-test) is widely used in hypothesis testing when you want to test claims about population means or proportions, especially when: The population standard deviation (œÉ) is known, or The sample size is large enough (usually n ‚â• 30) for the sample standard deviation to approximate the population‚Äôs.
### 16. How do you calculate a Z-score, and what does it represent ?
    - A Z-score tells you how many standard deviations a data point x is from the mean u of the distribution. 
### 17. What are point estimates and interval estimates in statistics ?
    - A point estimate is a single value calculated from sample data that serves as the best guess or approximation of an unknown population parameter (like the population mean, proportion, variance, etc.)
### 18. What is the significance of confidence intervals in statistical analysis ?
    - A confidence interval gives a range of plausible values for a population parameter (like a mean or proportion), calculated from sample data, along with a confidence level (e.g., 95%).
### 19. What is the relationship between a Z-score and a confidence interval ?
    - A Z-score is the critical value from the standard normal distribution used to calculate the width of a confidence interval. The confidence interval is constructed as the sample estimate plus or minus the Z-score multiplied by the standard error, defining the range where the true population parameter is likely to lie with a specified confidence level.
### 20. How are Z-scores used to compare different distributions ?
    - Z-scores allow you to standardize values from different distributions by converting them into a common scale with mean 0 and standard deviation 1. This way, you can directly compare how extreme or typical a value is relative to its own distribution, even if the original distributions have different means and spreads. Essentially, Z-scores turn raw scores into a universal measure of relative position across different datasets.
### 21. What are the assumptions for applying the Central Limit Theorem ?
    - Here are the key assumptions for applying the Central Limit Theorem (CLT):
    1. Independence: The sampled observations must be independent of each other (e.g., random sampling).
    2. Identically Distributed: The samples come from the same population distribution (i.i.d. ‚Äì independent and identically distributed).
    3. Sample Size: The sample size should be sufficiently large (usually n greater than equal to 30) for the theorem to hold well, especially if the original population is not normal.
    4. Finite Variance: The population from which samples are drawn should have a finite mean and variance.
### 22. What is the concept of expected value in a probability distribution ?
    - The expected value (or mean) of a probability distribution is the long-run average or weighted average of all possible values that a random variable can take, weighted by their probabilities.
### 23. How does a probability distribution relate to the expected outcome of a random variable?
    - A probability distribution describes all possible outcomes of a random variable and their associated probabilities. The expected outcome (or expected value) is calculated from this distribution by taking a weighted average of all possible values, using their probabilities as weights.
    - In other words, the expected value summarizes the probability distribution into a single number that represents the long-term average or mean outcome you can expect if the random process is repeated many times.
